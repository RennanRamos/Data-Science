```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('file:///C:/Users/renna/Downloads/Auto.csv')
df.head()
```

```python
lm = sm.OLS.from_formula('mpg ~ horsepower', df)
result = lm.fit()
result.summary()
```

```python
new = pd.DataFrame([[1, '98']], columns=['Intercept', 'horsepower'])
```

```python
result.predict(new)
```

```python
import matplotlib.pyplot as plt
df["horsepower"].replace({"?": "0"}, inplace=True)
df["horsepower"] = pd.to_numeric(df["horsepower"], downcast="float")
```

```python
df.plot.scatter('horsepower', 'mpg')
```

```python
df.corr()
```

```python
model = sm.OLS.from_formula('mpg ~ ' + '+'.join(df.columns.difference(['mpg', 'name'])), df)
result = model.fit()
result.summary()
```
